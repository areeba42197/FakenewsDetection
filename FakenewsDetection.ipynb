```python
import re

def worpot(text):
    # Convert text to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove digits
    text = re.sub(r'\d+', '', text)

    # Remove newline characters
    text = re.sub(r'\n', ' ', text)

     # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove extra whitespace (including tabs and multiple spaces)
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```
